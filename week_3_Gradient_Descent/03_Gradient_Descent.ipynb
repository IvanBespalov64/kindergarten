```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x : float) -> float:
    return x**3 + x**2 - 5*x
```

```python
x = np.linspace(-1, 2, 100)
y = f(x)
```

```python
plt.plot(x, y)
plt.xlim(-1, 2)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Давайте посчитаем аналитически минимум

x_min = 1
y_min = -3
```

```python
def grad(x : float) -> float:
    return 3*x**2 + 2*x - 5
```

```python
x_0 = -0.5
learning_rate = 0.2
tolerance = 1e-6

x_trace = [x_0]
```

```python
while True:
    x_old = x_trace[-1]
    x_new = x_old - learning_rate * grad(x_old)

    x_trace.append(x_new)

    if np.abs(x_new - x_old) < tolerance:
        break
```
