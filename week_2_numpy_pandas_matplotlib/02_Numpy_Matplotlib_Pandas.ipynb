## np.array()

```python
a = np.array([1, 2, 3])
print(a)
print(type(a))
```

## np.arange()

```python
b_nprange = np.arange(5)
b_range = range(5)

print(b_nprange)
print(type(b_nprange))

print(b_range)
print(type(b_range))
```

```python
power = 40
modulo = 10000
x1 = [(n ** power) % modulo for n in range(8)]
x2 = [(n ** power) % modulo for n in np.arange(8)]
print(x1, ' - correct')
print(x2, ' - incorrect')
```

## np.ones(), np.zeros(), np.eye()

```python
a_ones = np.ones((3, 1))
a_zeros = np.zeros((3, 3))
a_eye = np.eye(5)

print(a_ones, '\n')
print(a_zeros, '\n')
print(a_eye)
```

# Matplotlib

https://matplotlib.org/stable/

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3])
y = np.array([1, 2, 3])

plt.plot(x, y, color='red')
```

image.png

```python
plt.scatter(x, y)
```

```python
a = np.random.random((100, 100))
a
```

```python
plt.imshow(a)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(8, 5))
axs[0].scatter(x, y)
axs[1].plot(x, y)
```

```python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```
